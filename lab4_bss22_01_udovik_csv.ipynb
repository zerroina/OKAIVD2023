# 1.3 Работа с данными

В импортированном DataFrame необходимо создать новые столбцы, которые в каждой строке (на каждый день измерений) содержат информацию о минимальном, медианном и максимальном (min, median, max) значениях по 10 установкам.

Будьте внимательны используя функции min() и аналогичные: применяйте только с столбцам с данными (10шт), контролируйте параметр axis - вычислять по строкам, а не столбцам.

```python
file_csv.describe()
```

```python
cols = ['unit_01', 'unit_02', 'unit_03', 'unit_04', 'unit_05', 'unit_06', 'unit_07', 'unit_08', 'unit_09', 'unit_10']
```

```python
file_csv["min"] = file_csv[cols].min(axis=1)
file_csv["median"] = file_csv[cols].median(axis=1)
file_csv["max"] = file_csv[cols].max(axis=1)
```

```python
file_csv.head()
```

```python
file_csv.tail()
```

Последний столбец должен содержать данные на каждый день измерений в следующем формате:

МО/СКО (mean/std), т.е. математическое ожидание и стандартное отклонение. Поскольку это не одно число, а строка включающая 2 числа и разделяющий их символ « / », то нужно сформировать данные типа string. После применения функций вычисления МО и СКО, необходимо округлить значения до 2х знаков после запятой.

```python
mean_v = file_csv.mean(axis=1, numeric_only=True).round(2)
std_v = file_csv.std(axis=1, numeric_only=True).round(2)
list = []
for i in range(file_csv.shape[0]):
  list.append(str(mean_v[i])+"/"+str(std_v[i]))
file_csv["МО/СКО"] = list
```

```python
file_csv.head()
```

```python
file_csv.tail()
```

Наконец, необходимо добавить в конец новую строку с именем average annual, которая содержит средние значения по всем столбцам. За исключением последнего там строки, и результат будет не определен.

```python
# Выбор только числовых столбцов
import pandas as pd
num_cols = file_csv.select_dtypes(include=['float64', 'int64']).columns.tolist()

# Вычисление средних значений для числовых столбцов
mean_dict = {}
for column in num_cols:
  if column != "МО/СКО":
    mean_dict[column] = file_csv[column].mean()

# создаем новый DataFrame со средними значениями
mean_row = pd.DataFrame(mean_dict, index=['average annual']).round(2).astype(str)

# Добавление строки в DataFrame
file_csv = pd.concat([file_csv, mean_row])
```
